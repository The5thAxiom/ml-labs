```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../test_MachineLearningCVE.csv", skipinitialspace=True)
```

```python
print(df.head)
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, train_size=0.7, random_state=1)
```

```python
df_train.describe()
```

```python
df_test.describe()
```

```python
from sklearn.preprocessing import Normalizer, MinMaxScaler

X_train = df_train.iloc[:, 0:-2].to_numpy()
X_test = df_test.iloc[:, 0:-2].to_numpy()

norm = Normalizer()
norm.fit(X_train)
X_train = norm.transform(X_train)
X_test = norm.transform(X_test)

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

y_train = df_train.iloc[:, -1].to_numpy()
y_test = df_test.iloc[:, -1].to_numpy()

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
for i, feature in enumerate(df.columns[:-2]):
    print(f"{i}: {feature}")
    plt.plot(X_train[:, i])
    plt.title
    plt.show()
```

## Feature Extraction Using Principle Component Analysis
### Visualizing the Data
- To visualize the data succintly, 

```python
from sklearn.decomposition import PCA
```

```python
pca2d = PCA(n_components=2)
pcs = pca2d.fit_transform(X_train)

print(pcs)
```

```python
df2 = pd.DataFrame(
    data=pcs,
    columns=['PC1', 'PC2']
)
df2 = pd.concat(
    [df2, df[['Label']]],
    axis=1
)
df2.describe()
```

```python
from plotnine import ggplot, aes, geom_point

benign_plot = ggplot(
    df2[df2['Label'] == 0]
)+ aes(
    x='PC1',
    y='PC2'
) + geom_point()

malicious_plot = ggplot(
    df2[df2['Label'] != 0]
)+ aes(
    x='PC1',
    y='PC2',
    color='Label'
) + geom_point()
```

```python
benign_plot
```

```python
malicious_plot
```

- As the dataset is too large, transforming it to 2D doesn't give us a lot of inferences.
- Thus, a lot of variance is lost by such a drastic reduction.
- To train the dataset, we will reduce it such that 95% of the variance of the dataset is retained

```python
pca = PCA(.95)
pca.fit(X_train)
X_train_2 = pca.transform(X_train)
X_test_2 = pca.transform(X_test)
```

```python
pd.DataFrame(X_train_2).describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
lr1 = LogisticRegression()
lr1.fit(X_train, y_train)
```

```python
preds1 = lr1.predict(X_test)
print(classification_report(y_test, preds1))
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train_2, y_train)
```

```python
preds2 = lr2.predict(X_test_2)
print(classification_report(y_test, preds2))
```
