```python
import numpy as np
```

## Perceptron
### The parts

```python
X = np.array([1, 0, 0])
W = np.array([1, 1, 1])
```

```python
def unit_activation(n):
   return 1 if n > 0 else 0
```

```python
out = W.T * X
print(out)
ans = print(unit_activation(np.sum(out)))
```

```python
if ans == X[1] or X[2]:
    print('correct')
else:
    print('wrong')
```

### Combining the parts

```python
x0 = -1

thresh = -1
w0 = -thresh
w1 = 1
w2 = 1
W = np.array([w0, w1, w2])

num_iter = 2
learning_rate = 1

print("Perceptron\n===")
print("      X       |      W      |  sum  | pred | real | isCorrect")
for i in range(num_iter):
    for x1 in [0, 1]:
        for x2 in [0, 1]:
            X = np.array([x0, x1, x2])
            sum = np.sum(W.T * X)
            pred = unit_activation(sum)
            
            y = x1 or x2

            print(f"{X} |   {W}   |    {sum}    |    {pred}    |    {x1 or x2}   |    {bool(y == pred)}")
            
            # correcting the weights:
            if y != pred:
                dW = learning_rate * (y - pred) * X[1:]
                print(dW)
                W[1:] += dW
```
